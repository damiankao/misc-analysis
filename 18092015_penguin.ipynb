Reads were aligned with BWA MEM using the following command:

    bwa mem -t 32 -L 100,100 Aptenodytes_forsteri.scaf.fa EPADC001.1.format.fq.gz EPADC001.2.format.fq.gz | samtools view -b - > EPADC001.bam
    
Output BAM files were then filtered for unique alignments that have at most 5 mismatches and 2 indels. Using CIGAR string, MD and XA tag from the BWA-MEM output, we can filter for unique/mismatch/indels.

CIGAR string gives information on indels, MD gives information on mismatches, and the XA tag gives information on other references this read also maps to.

The following script is used to filter the BAM files:

    #! /usr/bin/env python
    import sys
    import re

    def parseMD(s):
        i = 0
        for c in s:
            if not c.isdigit():
                i += 1
        return i

    def parseCIGAR(s):
        match = re.findall(r'(\d+)(\w)', s)
        return (sum([int(x[0]) for x in match if x[1] == 'S']),sum([int(x[0]) for x in match if x[1] == 'I']), sum([int(x[0]) for x in match if x[1] == 'D']))

    pair = []
    for line in sys.stdin:
        if line[0] == '@':
            print line.strip()
        else:
            pair.append(line.strip().split())

            if len(pair) == 2:
                a_cigar = pair[0][5]
                b_cigar = pair[1][5]

                if a_cigar != '*' and b_cigar != '*':
                    a_tags = dict([(x[:2],x.split(':')[-1]) for x in pair[0][11:]])
                    b_tags = dict([(x[:2],x.split(':')[-1]) for x in pair[1][11:]])

                    if not a_tags.has_key('XA') and not b_tags.has_key('XA'):
                        a_clip, a_ins, a_del = parseCIGAR(a_cigar)
                        b_clip, b_ins, b_del = parseCIGAR(b_cigar)

                        if a_clip == 0 and b_clip == 0:
                            a_mm = parseMD(a_tags['MD']) - a_del
                            b_mm = parseMD(b_tags['MD']) - b_del
                            a_indel = a_ins + a_del
                            b_indel = b_ins + b_del

                            if a_mm <= 5 and b_mm <= 5 and a_indel <= 2 and b_indel <= 2:
                                print '\t'.join(pair[0])

                pair = []

```python
prefix = '/data/ngs/penguin_clean/stat/'
fs = os.listdir(prefix)
statTable = ListTable()

for f in fs:
    name = f.split('.')[0]
    stats = open(prefix + f).read().strip().split()
    total = int(stats[0])
    filtered = int(stats[-1])
    statTable.append([name,commas(total),commas(filtered),float(filtered) / total * 100])
statTable.sort(key = lambda x : x[-1])
```
