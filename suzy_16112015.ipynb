### Data trimming

```python
prefix = '/data/ngs/suzy/'
fs = [x for x in os.listdir('/data/ngs/suzy/') if x.split('.')[-1] == 'gz']
```

```python
fs = list(set(['_'.join(x.split('_')[:-1]) for x in fs]))
```

```python
trimCmdPEA = 'java -jar /home/share/software/Trimmomatic-0.32/trimmomatic-0.32.jar PE -threads 6 '
trimCmdB = ' LEADING:__QUAL__ TRAILING:__QUAL__ MINLEN:50 ' + \
'ILLUMINACLIP:/home/share/software/Trimmomatic-0.32/adapters/adapters.fa:2:30:12:1:false TOPHRED33'
trimCmds = ''

inPrefix = '/data/ngs/suzy/'
outPrefix = '/data/ngs/suzy/trim/'
for lib in fs:
    a = lib + '_1'
    b = lib + '_2'
    phred = detectPhred(inPrefix + a + '.fastq.gz', 5000)
    
    if phred == 'Sanger':
        phred = '-phred33 '
    else:
        phred = '-phred64 '
        
    #paired end
    trimCmds += trimCmdPEA + phred + inPrefix + a + '.fastq.gz ' + inPrefix + b + '.fastq.gz ' + \
    outPrefix + a + '.paired.trimmed.fastq.gz ' + outPrefix + a + '.unpaired.trimmed.fastq.gz ' + \
    outPrefix + b + '.paired.trimmed.fastq.gz ' + outPrefix + b + '.unpaired.trimmed.fastq.gz' + trimCmdB + ' 2> ' + lib + '.trim\n'
qTrimCmds = trimCmds.replace('__QUAL__','20')
HTML(textarea('Trimmomatic bash commands at quality 20, min length 50',qTrimCmds))
```

```python
sampleIndex = dict([x.split('\t') for x in open(prefix + 'samples.index').read().strip().split('\n')])
```

```python
SABP = 2820454
EFBP = 2739625
```

```python
readStats = dict([(x.split()[0],map(int,x.split()[1:])) for x in open('/data/ngs/suzy/reads.stats')])
```

```python
readStatList = ListTable()
for lib in fs:
    pairA = readStats[lib + '_1.fastq.gz']
    pairB = readStats[lib + '_2.fastq.gz']
    trimmedPA = readStats[lib + '_1.paired.trimmed.fastq.gz']
    trimmedPB = readStats[lib + '_2.paired.trimmed.fastq.gz']
    trimmedUA = readStats[lib + '_1.unpaired.trimmed.fastq.gz']
    trimmedUB = readStats[lib + '_2.unpaired.trimmed.fastq.gz']
    
    ref = 'EF'
    refCov = 0
    sumtrimmedBP = trimmedPA[1] + trimmedPB[1]
    if sampleIndex[lib][0].lower() == 's':
        ref = 'SA'
        refCov = sumtrimmedBP / float(SABP)
    else:
        refCov = sumtrimmedBP / float(EFBP)
        
    readStatList.append([lib, ref, commas(pairA[0]),commas(pairA[1]),commas(pairB[0]),commas(pairB[1]),\
                         trimmedPA[1] / float(pairA[1]) * 100,trimmedPB[1] / float(pairB[1]) * 100,\
                         refCov])
    
readStatList.sort(key = lambda x : x[0])
readStatList.insert(0,['Sample','ref','# reads A','# bases A','# reads B','# bases B',\
                       '% bases A after trimming','% bases B after trimming','Ref coverage'])
```

### Data mapping

Using bwa mem to map:

    bwa mem -R "@RG\tReadGroupID:\tSM:Sample' + lib + '\\tSM:SA" SA.fa libA libB > lib.sam 2> map.err
    


```python
cmds = ''
for lib in fs:
    aPath = prefix + 'trim/' + lib + "_1.paired.trimmed.fastq.gz" 
    bPath = prefix + 'trim/' + lib + "_2.paired.trimmed.fastq.gz"
    
    if sampleIndex[lib][0].lower() == 's':
        cmd = 'bwa mem -M -R "@RG\\tID:' + lib + '\\tSM:' + lib + '" SA.fa ' + aPath + ' ' + bPath + ' > ' + lib + '.sam 2> ' + lib + '.err'
        cmds += cmd + '\n'
    elif sampleIndex[lib][0].lower() == 'e':
        cmd = 'bwa mem -M -R "@RG\\tID:' + lib + '\\tSM:' + lib + '" EF.fa ' + aPath + ' ' + bPath + ' > ' + lib + '.sam 2> ' + lib + '.err'
        cmds += cmd + '\n'
HTML(textarea('Mapping commands',cmds))
```

### Post mapping processing

Sort:

    parallel echo "samtools sort -m 1G -O bam -T {.} -@ 32 {} \> {.}.sorted.bam" ::: *.sam > sort.sh
        
MarkDuplicates:

    parallel echo "java -Xmx20g -jar /home/share/software/picard-tools-1.115/MarkDuplicates.jar VALIDATION_STRINGENCY=LENIENT INPUT={} OUTPUT={.}.markdup.bam METRICS_FILE=File={.}.metrics" ::: *.sorted.bam
    
Call variants on single samples:

        java -Xmx30g -jar GenomeAnalysisTK.jar -T HaplotypeCaller --variant_index_type LINEAR --variant_index_parameter 128000 -ploidy 1 --emitRefConfidence GVCF -R EF.fa -I lib.sorted.markdup.bam -o lib.gvcf

```python
cmds = ''
for lib in fs:
    path = prefix + 'genome/' + lib + '.sorted.markdup.bam'
    
    if sampleIndex[lib][0].lower() == 's':
        cmd = 'java -Xmx30g -jar /home/share/software/GATK/GenomeAnalysisTK.jar -T HaplotypeCaller --variant_index_type LINEAR --variant_index_parameter 128000 -ploidy 1 --emitRefConfidence GVCF -R SA.fa -I ' + path + ' -o ' + lib + '.gvcf'
        cmds += cmd + '\n'
    elif sampleIndex[lib][0].lower() == 'e':
        cmd = 'java -Xmx30g -jar /home/share/software/GATK/GenomeAnalysisTK.jar -T HaplotypeCaller --variant_index_type LINEAR --variant_index_parameter 128000 -ploidy 1 --emitRefConfidence GVCF -R EF.fa -I ' + path + ' -o ' + lib + '.gvcf'
        cmds += cmd + '\n'
HTML(textarea('haplotype caller commands',cmds))
```
