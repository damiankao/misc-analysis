```python
data = [' '.join(x.split()[1:]).split('}{') for x in data.split('\n') if x.strip()[0] != '#']
```

```python
from collections import defaultdict

nodes = defaultdict(list)
formatted = []

for row in data:
    row[0] = row[0][1:]
    val = row[-1].split('}')[1].split('=')[-1][:-1]
    row[-1] = row[-1].split('}')[0]
    newRow = []
    for item in row:
        item = item[1:-1]
        if item != "":
            newRow.append(item)
    for i in range(len(newRow) - 1):
        nodes[newRow[i]].append(newRow[i + 1])
    nodes[newRow[-1]].append(val)
```

```python
for key in nodes.keys():
    nodes[key] = list(set(nodes[key]))
```

```python
parsed = ''
for k,v in nodes.items():
    parsed += k + '\t' + '\t'.join(v) + '\n'
HTML(textarea('first column is key, rest of the columns are children nodes',parsed))
```
