## SNP analysis of penguin population

Reads were gathered from 180 individuals. According to the barcoding of libraries, there are 116 'EP' (emperor penguin) individuals and 64 'KP' (king penguin) individuals. 

A total of 8 pair-end libraries were generated with 3 paired .fastq (6 total) each. 

Individual barcode sequences seems to be inline with the read sequences instead of in the header. Info for de-multiplexing the reads is in the .pools files describing the barcode sequence and invidual IDs.

The following python script was used to de-multiplex the reads:

```python
HTML(textarea('de-multiplex script','''import sys,gzip

prefix = sys.argv[1]
barcodes = dict([(x.strip().split()[1], x.strip().split()[0]) for x in open(sys.argv[2],'r')])

outLibs = {}
libCounts = {}
discarded = 0
total = 0
for bc, lib in barcodes.items():
        outLibs[lib + "1"] = open(lib + '.' + prefix + '_1.fastq','w')
        outLibs[lib + "2"] = open(lib + '.' + prefix + '_2.fastq','w')
        libCounts[lib] = 0

fileA = gzip.open(sys.argv[3],'r')
fileB = gzip.open(sys.argv[4],'r')

def getEntry(stream):
        seq = ''
        header = ''
        for line in stream:
                header = line.strip()
                break

        if header != '':
                seq = stream.next().strip()
                stream.next()
                qual = stream.next().strip()
                return [header,seq,'+',qual]
        else:
                return -1

while 1:
        entryA = getEntry(fileA)
        entryB = getEntry(fileB)
        if entryA == -1:
                break

        pos_A = -1
        type_A = 'NA'
        for bc in barcodes.keys():
                pos_A = entryA[1].find(bc)
                if pos_A != -1:
                        type_A = barcodes[bc]
                        break

        if pos_A != -1 and pos_A < 5:
                entryA[1] = entryA[1][pos_A + 8:]
                entryA[3] = entryA[4][pos_A + 8:]
                outLibs[type_A + '1'].write('\\n'.join(entryA) + '\\n')
                outLibs[type_A + '2'].write('\\n'.join(entryB) + '\\n')
                libCounts[type_A] += 1
        else:
                discarded += 1

        total += 1

print '\\n'.join([k + '\\t' + str(v) for k,v in libCounts.items()])
print 'discarded\t' + str(discarded)
print 'total\t' + str(total)
'''))
```

The script takes in a pair of .fastq files and look for exact matches to the barcode. The start of the barcode sequence must be found within the first 5 base-pairs of the forward read to be considered. Anything that doesn't match is discarded. Use the script by:

    python deplex.py <prefix> <barcode file> <forward fastq> <reverse fastq> > deplex.log

Not sure if this is the optimal method of de-multiplexing reads. The Stacks package can also perform this. It might do something more sophisticated like including reads with barcode off by a base or two.

### De-multiplex results

```python
metas = [f for f in os.listdir(prefix) if f.split('.')[-1] == 'meta']

inds = defaultdict(int)
total = 0
discarded = 0
for meta in metas:
    inFile = open(prefix + meta)
    for line in inFile:
        cols = line.strip().split()
        if cols[0] == 'total':
            total += int(cols[1])
        elif cols[0] == 'discarded':
            discarded += int(cols[1])
        else:
            inds[cols[0]] += int(cols[1])
        
    inFile.close()
```

### Trimming reads

Reads for each individual were trimmed for quality and possible adapter sequence that wasn't properly removed. Trimmomatic 0.32 was used for trimming with the following command:

    java -jar trimmomatic-0.32.jar PE -threads 6 -phred33 1.fastq 2.fastq 1.p.fastq 1.up.fastq 2.p.fastq 2.up.fastq LEADING:20 TRAILING:20 MINLEN:50 ILLUMINACLIP:adapters.fa:2:30:12:1:false TOPHRED33


```python
trimLog = open(prefix + 'trim.err')
statTable = ListTable()
lib = ''
stat = []
for line in trimLog:
    if line.find('TrimmomaticPE: Started') != -1:
        if lib != '':
            statTable.append([lib,commas(stat[0]),commas(stat[1]),str(100 * stat[1] / float(stat[0]))])
        meta = line.strip().split()
        lib = meta[7].split('/')[-1].split('_')[0]
        stat = []
    elif line.find('Input Read Pairs:') != -1:
        meta = line.strip().split()
        stat.append(int(meta[3]))
        stat.append(int(meta[6]))
statTable.sort(key = lambda x : int(''.join(x[1].split(','))), reverse = True)
statTable.insert(0,['Individual','Total reads','Reads kept','% reads kept'])
```

```python
htmlString = 'Libraries and number of reads after trimming<div style="overflow:auto;width:90%;height:500px;border:1px solid black;padding:20px;">'
htmlString += statTable._repr_html_()
htmlString += '</div>'
HTML(htmlString)
```

### Mapping reads to EP genome

Emperor Penguin genome (release 2011-07-06) was downloaded from:
    
    http://gigadb.org/dataset/100005
    ftp://climb.genomics.cn/pub/10.5524/100001_101000/100005/Aptenodytes_forsteri.scaf.fa.gz
    
BWA was used to index the genome for mapping. BWA was then used to align to the genome:

    bwa mem -t 32 Aptenodytes_forsteri.scaf.fa lib_1.fastq lib_2.fastq | samtools view -b - > lib.bam
    
Samtools was then used to add mate information and sort:
    
    samtools fixmate -O bam in.bam out.bam
    samtools sort -O bam -T tempname in.bam > out.bam
    
Picardtools was used to remove duplicates:

    java -Xmx10g -jar MarkDuplicates.jar VALIDATION_STRINGENCY=LENIENT INPUT=in.bam OUTPUT=out.bam METRICS_FILE=file.metric TMP_DIR=tmp MAX_FILE_HANDLES_FOR_READ_ENDS_MAP=20 REMOVE_DUPLICATES=true

```python
prefix = '/data/ngs/penguin/dup/'
processedTable = ListTable()
for row in statTable[1:]:
    path = prefix + row[0] + '.fixmate.sorted.dup.metric'
    metricFile = open(path)
    readNum = int(''.join(row[2].split(',')))
    pairsMapped = 0
    pairsDupped = 0
    for line in metricFile:
        if line.find('Unknown Library') != -1:
            meta = line.strip().split('\t')
            pairsMapped = int(meta[2])
            pairsDupped = float(meta[-2]) * 100
            break
    processedTable.append([row[0],commas(readNum),commas(pairsMapped),pairsDupped])
    metricFile.close()
processedTable.insert(0,['Library','Reads','Reads mapped','% duplicates'])
```

```python
htmlString = 'Libraries and mapped/duplicates<div style="overflow:auto;width:90%;height:500px;border:1px solid black;padding:20px;">'
htmlString += processedTable._repr_html_()
htmlString += '</div>'
HTML(htmlString)
```

I don't think it is appropriate to remove duplicated reads. Since this is rad-seq data, there will probably be a lot of similar reads generated that maps to the exact same position. There is a post of the GATK forum about this:

### Genome coverage per library

Genome coverage is first calculated for each library using bedtools' genome coverage module:

    genomeCoverageBed -ibam in.bam -bg > out.bg
    
I am using libraries where I have NOT removed duplicates with Picard Tools.
    
The output is bedgraph format. We can filter this bedgraph by different coverage thresholds by:

    awk '$4 >= 100' out.bg > out.filter100.bg
    
For each threshold, we can calculate coverage by adding up all intervals in the bed file:

    awk -F'\t' 'BEGIN{SUM=0}{ SUM+=$3-$2 }END{print SUM}' in.bg

```python
libCovDepth = defaultdict(list)
for i in range(0,101,10):
    path = '/data/ngs/penguin/cov/lib.' + str(i) + '.covDepth'
    inFile = open(path)
    for line in inFile:
        meta = line.strip().split()
        libCovDepth[meta[0]].append((int(meta[1]), float(meta[2])))
    inFile.close()
```

```python
covDepthTable = ListTable()
covs = []
depths = []
threshes = range(0,101,10)
for i in statTable[1:]:
    lib = i[0]
    covDepth = libCovDepth[lib]
    row = [lib]
    c = []
    d = []
    for item in covDepth:
        c.append(int(item[0]))
        d.append(float(item[1]))
        row.append(commas(item[0]) + '\n' + str(round(float(item[1]) * 100) / 100))
        
    covs.append(c)
    depths.append(d)
    covDepthTable.append(row)
covDepthTable.sort(key = lambda x : int(''.join(x[1].split()[0].split(','))), reverse = True)
covDepthTable.insert(0,['lib'] + threshes)
```

```python
htmlString = 'Sum covered regions/mean depth at increasing coverage thresholds (0-100)<div style="overflow:auto;width:90%;height:500px;border:1px solid black;padding:20px;">'
htmlString += covDepthTable._repr_html_()
htmlString += '</div>'
HTML(htmlString)
```

```python
covs = defaultdict(list)
depths = defaultdict(list)
threshes = range(0,101,10)

for i in statTable[1:]:
    lib = i[0]
    covDepth = libCovDepth[lib]

    for i, item in enumerate(covDepth):
        covs[threshes[i]].append(int(item[0]))
        depths[threshes[i]].append(float(item[1]))
```

```python
for i in threshes[:5]:
    s = pd.Series(covs[i])
    s[s < 1e8].plot(kind='kde',figsize=[12,6])
    
plt.legend(['coverage > 0','coverage >= 10','coverage >= 20','coverage >= 30','coverage >= 40'])
```

```python
for i in threshes[5:]:
    s = pd.Series(covs[i])
    s[s < 4e7].plot(kind='kde', figsize=[12,6])
    
plt.legend(['coverage >= 50','coverage >= 60','coverage >= 70','coverage >= 80','coverage >= 90','coverage >= 100'])
```

### Intersection of covered regions among all libraries

Libraries at each threshold (0,10,20,30..100) was first sorted by sum of covered regions. Serial intersections was then done from biggest to smallest using intersectBed -wo -a -b. The ouput was then parsed with a python script.

```python
intersections = {}
for i in range(0,101,10):
    path = '/data/ngs/penguin/cov/' + str(i) + '/'
    fs = [f for f in os.listdir(path) if f.split('.')[0] == 'intersection']
    covs = []
    for g in range(len(fs)):
        inFile = open('/data/ngs/penguin/cov/' + str(i) + '/intersection.' + str(g) + '.bed')
        total = 0
        for line in inFile:
            meta = line.strip().split()
            total += int(meta[2]) - int(meta[1])
        inFile.close()
        covs.append(total)
        intersections[i] = covs
```

```python
pd.DataFrame(intersections).plot(subplots=True, figsize=[12,50],sharex=False)
```
