# Telomere trace sequences

16,580,722 trace sequences for S. mediterranea in the trace archive. Sequences were clipped for low quality/vector contamination using clip info given in the trace archive.

1. Clipped sequences were searched for regex: (ccctaa)+ and (ttaggg)+. Greedy search.
2. For 'ccctaa' searches, potential repeat region is defined from start of sequence to the last position where 'ccctaa' is found. For 'ttaggg' searches, potential repeat region is defined from start of first found 'ttaggg' and end of the sequence.
3. A repeat percentage is found by dividing length of repeats by length of repeat region defined in step 2.
4. Filter for sequences with repeat percentage > 70%
5. Filter for sequences where after clipping potential repeat region, there are more than 20 base pairs leftover.
6. Use CAP3 to assemble these potential sub-telomeric regions

Below are the CAP3 assembled contigs. 

r+/r- or f+/f- gives orientation of the telomeres. r+/f- means 'ccctaa' is found in the beginning of the sequences, f+/r- means 'ttaggg' is found towards the end of the sequence. 

Number of traces are the number of trace reads that went into assembling this contig.

```python
prefix = '/home/share/projects/misc_analysis/sounak_25032015_telomere/'
```

```python
capFile = open(prefix + 'cap3.out')

header = ''
block = []

blocks = {}
for line in capFile:
    if line.find('*******************') != -1:
        header = ''.join(line.strip().split()[1:3])
        break

for line in capFile:
    if line.find('DETAILED DISPLAY') != -1:
        break
        
    if line.find('*******************') != -1:
        blocks[header] = block
        
        header = ''.join(line.strip().split()[1:3])
        block = []
    else:
        if line.strip() != '':
            block.append(line.strip().split()[0] + line.strip().split('is in')[0].split()[-1][-1])
            
blocks[header] = block
```

```python
faFile = open(prefix + 'telo.potentialSubtelo.fa.cap.contigs')
from Bio import SeqIO

faString = ''
for record in SeqIO.parse(faFile,'fasta'):
    block = blocks[record.id]
    orient = ','.join(list(set([x.split('_')[1] + x[-1] for x in block])))
    faString += '>' + record.id + '\t' + orient + '\tnumber of trace:' + str(len(block)) + '\n'
    faString += str(record.seq) + '\n'
```
