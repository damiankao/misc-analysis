# Tudor sequence phylogenetic analysis

```python
prefix = '/home/share/projects/misc_analysis/anish_07012016/'

rawFa = [x for x in open(prefix + 'tudor_raw.fa').read().strip().split('\n') if x.strip() != '']

entries = []
for i in range(0, len(rawFa),2):
    header = rawFa[i][1:].strip().replace('.','_')
    seq = rawFa[i + 1].strip()
    entries.append((header, seq))
    
HTML(textarea('raw sequences','\n'.join(['>' + x[0] + '\n' + x[1] for x in entries])))
```

```python
print 'Number of sequences:', len(entries)
```

Multiple alignment with mafft:

    mafft --auto tudor_formatted.fa > tudor_formatted.mafft.fa

```python
mafftFa = open(prefix + 'tudor_formatted.mafft.fa').read().strip()

HTML(textarea('aligned sequences',mafftFa))
```

Trim alignment with trimal:

    trimal -in tudor_formatted.mafft.fa -out tudor_formatted.mafft.trimal.fa -automated1

```python
trimmedFa = open(prefix + 'tudor_formatted.mafft.trimal.fa').read().strip()

HTML(textarea('trimmed sequences',trimmedFa))
```

Use fa2phylip script to convert to phylip format:

    sh fa2phylip.sh tudor_formatted.mafft.trimal.fa > tudor_formatted.mafft.trimal.phylip

```python
phylipData = open(prefix + 'tudor_formatted.mafft.trimal.phylip').read().strip()

HTML(textarea('phylip formatted',phylipData))
```
